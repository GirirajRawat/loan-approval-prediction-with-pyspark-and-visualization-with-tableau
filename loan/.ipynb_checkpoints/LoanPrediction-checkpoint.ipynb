```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Loan Prediction Dataset.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# find the null values
df.isnull().sum()
```

```python
# fill the missing values for numerical terms - mean
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
# # fill the missing values for categorical terms - mode
# df['Dependents'] = df['Dependents'].str.replace('+', '').astype(int)
# df['Gender'] = df["Gender"].str.replace('+', '').astype(int)

# Convert 'Dependents' column to string type
df['Dependents'] = df['Dependents'].astype(str)
# df['Gender'] = df['Gender'].astype(str)

# Replace '+' and convert to integer
df['Dependents'] = df['Dependents'].str.replace('+', '').astype(int)
# df['Gender'] = df["Gender"].str.replace('+', '').astype(int)
# df['Married'] = df['Married'].str.replace('+', '').astype(int)
```
